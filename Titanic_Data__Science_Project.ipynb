### DATA VISUALISATION

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Survive Rate
survive_rate = df['survived'].value_counts(normalize=True)* 100
survive_rate.plot(kind="bar")
plt.title("Survive Percentage")
plt.ylabel("Percetange")
plt.show()
```

###### Survive percentage:65%  of people survive

```python
sns.barplot(x='sex',y='survived',data=df)
plt.title('Survive rate by Gender')
plt.show()
```

####### Survive rateby gender : more female than male

```python
sns.boxplot(x='pclass', y='fare',data=df)
plt.title('Fare distribution by class')
plt.show()
```

########  class 1 people pay more on fare

### FEATURE ENGINEERING

```python
df['family_size'] = df['parch'] + df['sibsp']
sns.barplot(x='family_size', y='survived', data=df)
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
sns.histplot(data=df,x='age',hue='survived',palette=['red','yellow'])
plt.title("Age Distribution")
plt.show()
```

```python
survival=df['survived'].value_counts()
plt.pie( survival,labels=['Not Survived', 'survived'],autopct="%1.1f%%", colors=['blue','green'])
plt.title('Survive Percentage')
plt.show()
```
